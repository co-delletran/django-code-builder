```python
def read_file(path="read_txt.txt"):
    with open(path, "r") as f:
        result = f.read()
        f.close

    return result


def write_to_file(str_, path="writequery.txt", append=False):
    if append:
        with open(path, "a") as f:
            f.writelines("\n")
            f.writelines(str_)
            f.close
    else:
        with open(path, "w") as f:
            f.write(str_)
            f.close
```

```python
import_basic_model_snippet = '''
from django.db import models
from django.urls import reverse
from django.utils import timezone
from django.utils.translation import gettext_lazy as _
$UPDATE_IMPORTS

'''

import_uuid_snippet = '''
import uuid
import shortuuid'''

import_user_model_snippet = '''
from django.contrib.auth import get_user_model
USER = get_user_model()'''

import_signals_snippet = '''
from django.db.models.signals import post_save
from django.dispatch import receiver'''


import_generic_content_type_snippet = '''
from django.contrib.contenttypes.models import ContentType
from django.contrib.contenttypes.fields import GenericRelation
from django.contrib.contenttypes.fields import GenericForeignKey'''



class_snippet = '''
class $TABLE_NAME(models.Model):'''

char_attribute_snippet = '''
    $VAR_ATTRIB_NAME = models.CharField(max_length=64, blank=True, null=True)'''
    
textfield_attribute_snippet = '''
    $VAR_ATTRIB_NAME = models.TextField(max_length=512, blank=True, null=True)'''
    
int_attribute_snippet = '''
    $VAR_ATTRIB_NAME = models.IntegerField(blank=True, null=True)'''

bool_attribute_snippet = '''
    $VAR_ATTRIB_NAME = models.BooleanField(default=False)'''

email_attribute_snippet = '''
    $VAR_ATTRIB_NAME = models.EmailField('email address', max_length=64, unique=True)'''

positive_int_attribute_snippet = '''
    $VAR_ATTRIB_NAME = models.PositiveSmallIntegerField(
        verbose_name=_("Index/Order"),
        null=True,
        default=None
    )'''

datetime_attribute_snippet = '''
    $VAR_ATTRIB_NAME = models.DateTimeField(auto_now_add=False, auto_now=False, default=timezone.now)'''

varbinary_attribute_snippet = '''
    $VAR_ATTRIB_NAME = models.ImageField(
        upload_to=UPLOAD_TO,
        null=True,
        blank=True
    )'''

fk_attribute_snippet = '''
    $VAR_ATTRIB_NAME = models.ForeignKey(
        "$FK_ATTRIB_NAME",
        verbose_name=_("$FK_ATTRIB_NAME"),
        db_column='$PK_NAME',
        related_name='$REL_FK_ATTRIB_NAME',
        on_delete=models.CASCADE
        )'''

m2m_attribute_snippet = '''
    $VAR_ATTRIB_NAME = models.ManyToManyField(
        "$FK_ATTRIB_NAME",
        verbose_name=_("$FK_ATTRIB_NAME"),
        db_column='$PK_NAME',
        related_name='$REL_FK_ATTRIB_NAME',
        blank=True,
        )'''
        
content_type_attribute_snippet = '''
    $VAR_ATTRIB_NAME = models.ForeignKey(ContentType, null=True, blank=True)'''
    
content_object_attribute_snippet = '''
    $VAR_ATTRIB_NAME = GenericForeignKey('content_type','object_id')'''

class_meta_snippet = '''
    class Meta:
        db_table = 'db_$CLASS_NAME'
        ordering = ['-pk']
        verbose_name = "$VERBOSE_CLASS_NAME"
        verbose_name_plural = "$PLURAL_VERBOSE_CLASS_NAME"
    '''

str_fk_rep_method_snippet = '''

    def __str__(self):
        return f"\{self.$VAR_ATTRIB_NAME.$VAR_ATTRIB_NAME\}"
'''

str_rep_method_snippet = '''

    def __str__(self):
        return f"\{self.$VAR_ATTRIB_NAME}"
'''

absolute_url_method_snippet = '''
    def get_absolute_url(self):
        return reverse("$VAR_ATTRIB_NAME_detail", kwargs={"pk": self.pk})
'''

post_save_function_snippet = '''
@receiver(post_save, sender=$VAR_ATTRIB_NAME)
def post_save_$VAR_ATTRIB_NAME(sender, instance, created, *args, **kwargs):
    if created:
        profile = UserProfile(
            $VAR_ATTRIB_NAME=instance,
        )
        profile.save()

'''
```

```python
class CodeBuilder():
    def __init__(self, json_datamodel):
        self.json_datamodel = json_datamodel
        self.generated_models = []

    def generate_model(self):
        generated_models = []
        for table_idx, table in enumerate(self.json_datamodel):
            globals()[table['name']] = import_basic_model_snippet
            self.append_class(table['name'])
            self.append_class_meta(table['name'])
            datatypes = []

            for field_idx, field in enumerate(table['fields']):
                datatype = field['field'].get('type')
                datatypes.append(datatype)
                is_fk = field['field'].get('is_fk')
                source_id = field['field'].get('source_id') if field['field'].get(
                    'source_id') else field['field']['foreign_key'].get('source_id')
                attrib_name = field['field'].get('name')

                primary_key = None
                primary_table = self.get_primary_table(source_id)
                if primary_table:
                    primary_key = primary_table.get('name')

                _code_snippet = self.get_snippet_attrib_by_type(
                    datatype, attrib_name, is_fk, primary_key)
                if attrib_name:
                    globals()[table['name']] += _code_snippet

            self.append_methods(table['name'], attrib_name, datatype)

            if len(set(datatypes)) >= 2:
                generated_models.append(
                    {"name": table['name'], "code": globals()[table['name']]})

        return generated_models

    def append_class(self, table_name):
        if table_name:
            globals()[
                table_name] += self.replace_tags(class_snippet, table_name)

    def append_class_meta(self, table_name):
        if table_name:
            globals()[
                table_name] += self.replace_tags(class_meta_snippet, table_name)

    def append_methods(self, table_name, attrib_name, datatype):
        if attrib_name and datatype == 'char':
            globals()[
                table_name] += self.replace_tags(str_rep_method_snippet, attrib_name)
            globals()[
                table_name] += self.replace_tags(absolute_url_method_snippet, attrib_name)

    def get_primary_table(self, source_id):
        if source_id:
            for table_idx, table in enumerate(self.json_datamodel):
                if table['name']:
                    for field_idx, field in enumerate(table['fields']):
                        if field['field_group_id'] == source_id:
                            return field.get('field')

        return None

    def snippet_generator(self):
        pass

    def get_snippet_attrib_by_type(self, type: str, attrib_name: str, is_fk=False, primary_key=None) -> str:
        match type:
            case "char":
                return self.replace_tags(char_attribute_snippet, attrib_name)
            case "int":
                return self.replace_tags(int_attribute_snippet, attrib_name)
            case "date/time":
                return self.replace_tags(datetime_attribute_snippet, attrib_name)
            case "varbinary":
                return self.replace_tags(varbinary_attribute_snippet, attrib_name)
            case "bool":
                return self.replace_tags(bool_attribute_snippet, attrib_name)
            case "content_type":
                _snippet = self.replace_tags(
                    content_type_attribute_snippet, attrib_name)
                # _snippet =+ self.replace_tags(content_object_attribute_snippet, attrib_name)
                return _snippet
            case "email":
                return self.replace_tags(email_attribute_snippet, attrib_name)
            case "ps_int":
                return self.replace_tags(positive_int_attribute_snippet, attrib_name)

            case "":
                if is_fk:
                    _snippet = self.replace_tags(
                        fk_attribute_snippet, attrib_name, primary_key)
                    return _snippet
                return ''
            case _:
                return ''

    def replace_tags(self, snippet: str, value: str, primary_key: str = None) -> str:
        formatted_snippet = snippet
        formatted_snippet = formatted_snippet.replace(
            f"$TABLE_NAME", value.title())
        formatted_snippet = formatted_snippet.replace(
            f"$VAR_ATTRIB_NAME", value.lower())
        formatted_snippet = formatted_snippet.replace(
            f"$FK_ATTRIB_NAME", value.title().replace('_', ''))
        formatted_snippet = formatted_snippet.replace(
            f"$REL_FK_ATTRIB_NAME", value.lower())
        # formatted_snippet = formatted_snippet.replace(
        #     f"$PK_NAME", f"{value.lower()}_id")
        formatted_snippet = formatted_snippet.replace(
            f"$CLASS_NAME", f"{value.title()}")
        formatted_snippet = formatted_snippet.replace(
            f"$VERBOSE_CLASS_NAME", f"{value.title()}")
        formatted_snippet = formatted_snippet.replace(
            f"$PLURAL_VERBOSE_CLASS_NAME", f"{value.title()}s")

        # print(f"-- primary_key: {primary_key}")
        if primary_key:
            formatted_snippet = formatted_snippet.replace(
                f"$PK_NAME", primary_key.lower())

        return formatted_snippet
```
